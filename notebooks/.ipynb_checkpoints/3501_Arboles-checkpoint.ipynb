# Árboles de Decisión

En este caso práctico se pretende resolver un problema de malware en dispositivos android mediante e analisis del tráfico

## Imports

```python
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.preprocessing import RobustScaler
from sklearn.metrics import f1_score, confusion_matrix, recall_score, precision_score
from pandas import DataFrame
```

## Funciones Auxiliares

```python
def train_val_test_split(df, rstate=42, shuffle=True, stratify=None):
    strat = df[stratify] if stratify else None
    train_set, test_set = train_test_split(
        df, test_size=0.4, random_state=rstate, shuffle=shuffle, stratify=strat)
    strat = test_set[stratify] if stratify else None
    val_set, test_set = train_test_split(
        test_set, test_size=0.5, random_state=rstate, shuffle=shuffle, stratify=strat)
    return (train_set, val_set, test_set)
```

```python
def remove_labels(df, label_name):
    X = df.drop(label_name, axis=1)
    y = df[label_name].copy()
    return (X, y)
```

```python
def evaluate_result(y_pred, y, y_prep_pred, y_prep, metric):
    print(metric.__name__, "WITHOUT preparation:", metric(y_pred, y, average='weighted'))
    print(metric.__name__, "WITH preparation:", metric(y_prep_pred, y_prep, average='weighted'))
```

```python
# 3.- División del DataSet
train_set, val_set, test_set = train_val_test_split(X)
```

```python
X_train, y_train = remove_labels(train_set, 'calss')
X_val, y_val = remove_labels(val_set, 'calss')
X_test, y_test = remove_labels(test_set, 'calss')
```

## 4.- Escalar el DataSet

Es importante comprendeer que los árbole son algoritmos que **no requieren de preparación de datos**, concretamnete no requieren realizar el escalado o normalización. En este ejercicio se va a realizar escalado al DataSet y se van a comparar los resultados con el DataSet sin escalar. De esta forma se demuestra, como aplicar preprocesamientos y como el escalado puede incuso afectar el escalado del rendimeino del módelo.

```python
scaler = RobustScaler()
X_train_scaler = scaler.fit_transform(X_train)
```

```python
scaler = RobustScaler()
X_test_scaler = scaler.fit_transform(X_test)
```

```python
scaler = RobustScaler()
X_val_scaler = scaler.fit_transform(X_val)
```

```python
# Transformar un DataFrame de pandas
X_train_scaled = DataFrame(X_train_scaler, columns=X_train.columns, index=X_train.index)
X_train_scaled.head(10)
```

```python
X_train_scaled.describe()
```

## 5.- Árbol de decision

```python
from sklearn.tree import DecisionTreeClassifier

MAX_DEPTH = 20

# Módelo entrenado con el DataSet sin escalar
clf_tree = DecisionTreeClassifier(max_depth = MAX_DEPTH, random_state=42)
clf_tree.fit(X_train, y_train)
```

```python
# Modelo entrenado con el conjunto de datos 
clf_tree_scaled = DecisionTreeClassifier(max_depth=MAX_DEPTH, random_state=42)
clf_tree_scaled.fit(X_train_scaled, y_train)
```

Comenzar la prediccion con el propio DataSet, con el que se a entrenado el algoritmo(trainset), suele ser interezante para comprobar si se esta realizando el OverFitting.

```python
#Predecimos con el conjunto de datos de entrenamiento.
y_train_pred = clf_tree.predict(X_train)
y_train_pred_prep = clf_tree_scaled.predict(X_train_scaled)
```

```python
# Comparacion resultados entre escalado y sin escalar
evaluate_result(y_train_pred, y_train, y_train_pred_prep, y_train, f1_score)
                
               
```

```python
# Predecir el dataSet de validación
y_pred = clf_tree.predict(X_val)
y_pred_prep = clf_tree_scaled.predict(X_val)
```

```python
# Comparar los resultados entre el escalado y sin escalar
evaluate_result(y_pred, y_val, y_pred_prep, y_val, f1_score)
```

A partir de este punto, y, en función de los resultados que se han visto, nos quedamos con la opción de no escalar nuestros DataSet's

## 6.- Visualizando el limite de decisión

```python
# Reducir el numero de atributos del DataSet para la visualizarlo mejor.
X_train_reduced = X_train[['min_flowpktl', 'flow_fin']]
```

```python
# Generar un modelo con el DataSet reducido.
clf_tree_reduced = DecisionTreeClassifier(max_depth=2, random_state=42)
clf_tree_reduced.fit(X_train_reduced, y_train)
```

```python
# Representamos gráficamente el límite de decisión construido
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt
%matplotlib inline

def plot_decision_boundary(clf, X, y, plot_training=True, resolution=1000):
    mins = X.min(axis=0) - 1
    maxs = X.max(axis=0) + 1
    x1, x2 = np.meshgrid(np.linspace(mins[0], maxs[0], resolution),
                         np.linspace(mins[1], maxs[1], resolution))
    X_new = np.c_[x1.ravel(), x2.ravel()]
    y_pred = clf.predict(X_new).reshape(x1.shape)
    custom_cmap = ListedColormap(['#fafab0','#9898ff','#a0faa0'])
    plt.contourf(x1, x2, y_pred, alpha=0.3, cmap=custom_cmap)
    custom_cmap2 = ListedColormap(['#7d7d58','#4c4c7f','#507d50'])
    plt.contour(x1, x2, y_pred, cmap=custom_cmap2, alpha=0.8)
    if plot_training:
        plt.plot(X[:, 0][y==0], X[:, 1][y==0], "yo", label="normal")
        plt.plot(X[:, 0][y==1], X[:, 1][y==1], "bs", label="adware")
        plt.plot(X[:, 0][y==2], X[:, 1][y==2], "g^", label="malware")
        plt.axis([mins[0], maxs[0], mins[1], maxs[1]])               
    plt.xlabel('min_flowpktl', fontsize=14)
    plt.ylabel('flow_fin', fontsize=14, rotation=90)

plt.figure(figsize=(12, 6))
plot_decision_boundary(clf_tree_reduced, X_train_reduced.values, y_train)
plt.show()
```
